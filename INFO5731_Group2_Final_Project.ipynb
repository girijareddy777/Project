```python
# Deleting unwanted colunmns from the data.
final_data = file_3k[['Authors','Article Title','Language','Abstract','Addresses','Author Keywords','Cited Reference Count','Affiliations','Publication Year','Research Areas']]
final_data.head(10)
```

```python
# Select the first 100 rows
df_subset = final_data.head(100)

# Split authors, research areas, and affiliations
authors_and_areas = df_subset[['Authors', 'Research Areas', 'Affiliations']].dropna()

# Create a list to store unique authors with their corresponding research areas and affiliations
author_info_list = []

# Iterate through rows to populate the list
for index, row in authors_and_areas.iterrows():
    authors = row['Authors'].split(';')
    areas = row['Research Areas'].split(';')
    affiliations = row['Affiliations'].split(';')

    # Combine authors with their respective research areas and affiliations
    for author, affiliation in zip(authors, affiliations):
        author_info_list.append({'Author': author.strip(), 'Research Areas': areas, 'Affiliation': affiliation.strip()})

# Create a new DataFrame from the list
author_info_df = pd.DataFrame(author_info_list)

# Flatten the list of research areas
flat_areas = [area.strip() for sublist in author_info_df['Research Areas'] for area in sublist]

# Create a new DataFrame with authors, their corresponding research areas, and affiliations
flat_author_info_df = pd.DataFrame({'Author': author_info_df['Author'].repeat(author_info_df['Research Areas'].apply(len)),
                                     'Research Areas': flat_areas,
                                     'Affiliation': author_info_df['Affiliation'].repeat(author_info_df['Research Areas'].apply(len))})

# Plot a countplot to visualize the distribution of research areas per author and affiliation
plt.figure(figsize=(15, 8))
sns.countplot(y='Research Areas', hue='Affiliation', data=flat_author_info_df,
              order=flat_author_info_df['Research Areas'].value_counts().index)
plt.title('Distribution of Research Areas per Author and Affiliation (First 10 Articles)')
plt.xlabel('Count')
plt.ylabel('Research Areas')
plt.legend(title='Affiliation', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
from itertools import combinations

# Select relevant columns
authors_and_areas = final_data[['Article Title', 'Authors', 'Research Areas', 'Affiliations', 'Abstract']].dropna()

# Create a list to store unique combinations of article title, authors, research areas, affiliations, and abstracts
article_info_list = []

# Iterate through rows to populate the list
for _, row in authors_and_areas.iterrows():
    title = row['Article Title']
    authors = row['Authors'].split(';')
    areas = row['Research Areas'].split(';')
    abstracts = row['Abstract'].split(';')

    # Check if an article has 2 or more authors and authors are from different research areas
    if len(authors) >= 2 and len(set(areas)) >= 2:
        # Generate all possible combinations of authors
        author_combinations = list(combinations(authors, 2))

        # Append the combinations to the list, along with the article title, research areas, and one abstract
        article_info_list.extend([(title, area1, area2, abstract) for (area1, area2), abstract in zip(combinations(areas, 2), abstracts) for _, _ in author_combinations])

# Create a new DataFrame from the list
article_area = pd.DataFrame(article_info_list, columns=['Article Title', 'Research Area 1', 'Research Area 2', 'Abstract'])

# Group by article title and count unique author pairs
interdisciplinary_count = article_area.groupby(['Article Title','Research Area 1','Research Area 2','Abstract']).size().reset_index(name='Author Pair Count')

# Filter articles with 2 or more author pairs
interdisciplinary_count = interdisciplinary_count[interdisciplinary_count['Author Pair Count'] >= 2]

# Sort by the author pair count in ascending order
interdisciplinary_count_1 = interdisciplinary_count.sort_values(by='Author Pair Count', ascending=True)

# Print the resulting DataFrame

interdisciplinary_count_1.head(50)
```

```python
import string
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import pandas as pd
import nltk

# Download WordNet resource
nltk.download('wordnet')

# Sample DataFrame with the selected articles (replace this with your actual DataFrame)
selected_articles_df = selected_articles[['Research Area 1', 'Research Area 2', 'Article Title', 'Abstract']]
# Function to remove special characters
def remove_special_characters(text):
    return re.sub(r'[^A-Za-z0-9 ]+', '', text)

# Function to tokenize the text
def tokenize_text(text):
    return word_tokenize(text.lower())

# Function to remove stopwords
def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    return [token for token in tokens if token not in stop_words]

# Function for stemming
def perform_stemming(tokens):
    stemmer = PorterStemmer()
    return [stemmer.stem(token) for token in tokens]

# Function for lemmatization
def perform_lemmatization(tokens):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(token) for token in tokens]

# Apply text preprocessing to each abstract in the selected articles DataFrame
for index, row in selected_articles_df.iterrows():
    abstract = row['Abstract']

    # Remove special characters
    cleaned_abstract = remove_special_characters(abstract)

    # Tokenize the text
    tokens = tokenize_text(cleaned_abstract)

    # Remove stopwords
    tokens_without_stopwords = remove_stopwords(tokens)

    # Perform stemming
    stemmed_tokens = perform_stemming(tokens_without_stopwords)

    # Perform lemmatization
    lemmatized_tokens = perform_lemmatization(tokens_without_stopwords)

    # Update the 'Abstract' column with the preprocessed text
    selected_articles_df.at[index, 'Abstract'] = ' '.join(lemmatized_tokens)

# Display the DataFrame with preprocessed abstracts
selected_articles_df
```

# Topic Modeling -LDA 

```python
import nltk
from nltk.tokenize import word_tokenize
import gensim.corpora as corpora

nltk.download('punkt')

# Tokenize each abstract in final_data['abstract_new']
tokenized_abstracts = [word_tokenize(abstract) for abstract in selected_articles_df['Abstract']]

# Create Dictionary
id2word = corpora.Dictionary(tokenized_abstracts)

# Term Document Frequency
corpus = [id2word.doc2bow(abstract) for abstract in tokenized_abstracts]

# View
print(corpus[:1][0][:30])
```

```python
import gensim
from gensim.utils import simple_preprocess
from pprint import pprint
# number of topics
num_topics = 15
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 15 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```
